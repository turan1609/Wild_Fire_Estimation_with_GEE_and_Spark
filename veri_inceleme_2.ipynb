```python
import findspark
findspark.init()
from pyspark.sql import SparkSession
from pyspark.sql.functions import col, month
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Spark'Ä± BaÅŸlat
spark = SparkSession.builder.appName("ALAZ_EDA").master("local[3]").getOrCreate()

# Veriyi Oku
df = spark.read.parquet("../data/processed/ALAZ_Final_Dataset_Cleaned.parquet")

# Veriye "Ay" bilgisini ekleyelim (Mevsimsellik analizi iÃ§in lazÄ±m)
df = df.withColumn("ay", month(col("tarih")))

# Veriyi Pandas'a Ã§evir (370k satÄ±r RAM'e sÄ±ÄŸar, korkma)
print("ðŸ“¦ Veri Pandas formatÄ±na dÃ¶nÃ¼ÅŸtÃ¼rÃ¼lÃ¼yor...")
pdf = df.toPandas()

print("âœ… Veri hazÄ±r! Grafik Ã§izimine baÅŸlÄ±yoruz.")

# Grafik AyarlarÄ± (Profesyonel GÃ¶rÃ¼nÃ¼m Ä°Ã§in)
sns.set_style("whitegrid")
plt.rcParams.update({'font.size': 12})
```

```python
plt.figure(figsize=(14, 10))

# Sadece sayÄ±sal sÃ¼tunlarÄ± alalÄ±m
cols_for_corr = ['yukselti', 'egim', 'baki', 'sicaklik', 'ruzgar_hizi', 'bagil_nem', 'yagis', 'toprak_sicakligi', 'NDVI', 'label']
corr_matrix = pdf[cols_for_corr].corr()

# IsÄ± HaritasÄ±
sns.heatmap(corr_matrix, annot=True, fmt=".2f", cmap='coolwarm', linewidths=0.5, vmin=-1, vmax=1)
plt.title("Ã–znitelikler ArasÄ± Korelasyon Matrisi (Ä°liÅŸki HaritasÄ±)", fontsize=16)
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(18, 6))

# 1. Aylara GÃ¶re DaÄŸÄ±lÄ±m (Sadece YangÄ±n Olanlar)
fire_only = pdf[pdf['label'] == 1]
sns.countplot(x='ay', data=fire_only, palette='magma', ax=ax[0])
ax[0].set_title("Aylara GÃ¶re YangÄ±n SayÄ±sÄ± (Mevsimsellik)")
ax[0].set_xlabel("Ay")
ax[0].set_ylabel("YangÄ±n SayÄ±sÄ±")

# 2. Saate GÃ¶re DaÄŸÄ±lÄ±m
sns.countplot(x='saat', data=fire_only, palette='Reds_r', ax=ax[1])
ax[1].set_title("GÃ¼nÃ¼n Saatlerine GÃ¶re YangÄ±n SayÄ±sÄ± (Ä°nsan Etkisi)")
ax[1].set_xlabel("Saat")
ax[1].set_ylabel("YangÄ±n SayÄ±sÄ±")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# Veri Ã§ok yoÄŸun olduÄŸu iÃ§in 5000 tane rastgele Ã¶rnek alÄ±p Ã§izelim (Daha hÄ±zlÄ± ve net olur)
sample_pdf = pdf.sample(n=5000, random_state=42)

sns.scatterplot(
    data=sample_pdf, 
    x='sicaklik', 
    y='bagil_nem', 
    hue='label', 
    palette={0: 'green', 1: 'red'}, # 0: YeÅŸil (GÃ¼venli), 1: KÄ±rmÄ±zÄ± (YangÄ±n)
    alpha=0.6,
    s=30
)

plt.title("SÄ±caklÄ±k ve Nem Ä°liÅŸkisinin YangÄ±na Etkisi")
plt.xlabel("Hava SÄ±caklÄ±ÄŸÄ± (Â°C)")
plt.ylabel("BaÄŸÄ±l Nem (%)")
plt.legend(title='YangÄ±n Durumu', labels=['Yok', 'Var'])
plt.axhline(y=30, color='blue', linestyle='--', label='Kritik Nem EÅŸiÄŸi (%30)') # Kritik nem Ã§izgisi
plt.show()
```

```python
fig, ax = plt.subplots(1, 2, figsize=(16, 6))

# YÃ¼kselti Analizi
sns.violinplot(x='label', y='yukselti', data=pdf, palette="Set2", ax=ax[0], split=True)
ax[0].set_title("YÃ¼kseltiye GÃ¶re YangÄ±n DaÄŸÄ±lÄ±mÄ±")
ax[0].set_xticklabels(['YangÄ±n Yok', 'YangÄ±n Var'])

# EÄŸim Analizi
sns.boxplot(x='label', y='egim', data=pdf, palette="Set3", ax=ax[1])
ax[1].set_title("EÄŸime GÃ¶re YangÄ±n DaÄŸÄ±lÄ±mÄ±")
ax[1].set_xticklabels(['YangÄ±n Yok', 'YangÄ±n Var'])

plt.show()
```

```python
plt.figure(figsize=(10, 6))

sns.histplot(
    data=pdf, 
    x="NDVI", 
    hue="label", 
    element="step", 
    stat="density", 
    common_norm=False, 
    palette={0: 'green', 1: 'red'},
    bins=50
)

plt.title("Bitki Ã–rtÃ¼sÃ¼ YoÄŸunluÄŸu (NDVI) DaÄŸÄ±lÄ±mÄ±")
plt.xlabel("NDVI DeÄŸeri (-1 ile 1 arasÄ±)")
plt.ylabel("YoÄŸunluk")
plt.xlim(0, 1) # Sadece bitki olan kÄ±sma odaklanalÄ±m
plt.show()
```

```python
plt.figure(figsize=(10, 6))

# RÃ¼zgar hÄ±zÄ± genelde dÃ¼ÅŸÃ¼k olduÄŸu iÃ§in logaritmik bakmak yerine direkt KDE (YoÄŸunluk) Ã§izelim
sns.kdeplot(data=pdf[pdf['label']==0]['ruzgar_hizi'], shade=True, color="green", label="YangÄ±n Yok")
sns.kdeplot(data=pdf[pdf['label']==1]['ruzgar_hizi'], shade=True, color="red", label="YangÄ±n Var")

plt.title("RÃ¼zgar HÄ±zÄ±nÄ±n YangÄ±n Ãœzerindeki Etkisi")
plt.xlabel("RÃ¼zgar HÄ±zÄ± (m/s)")
plt.ylabel("Frekans")
plt.legend()
plt.show()
```
